```python
import pandas as pd
import numpy as np

# Функция для перемещения последнего ненулевого значения в 63-й столбец
def move_last_non_nan_to_end(row):
    # Преобразуем строку в список
    row_values = row.tolist()

    # Находим индекс последнего ненулевого значения
    last_non_nan_index = next((i for i in range(len(row_values) - 1, -1, -1) if not pd.isna(row_values[i])), None)

    if last_non_nan_index is not None and last_non_nan_index != 62:  # 62-й индекс = 63-й столбец
        # Сохраняем последнее ненулевое значение
        last_non_nan_value = row_values[last_non_nan_index]

        # Убираем значение из текущего места
        row_values[last_non_nan_index] = np.nan

        # Устанавливаем его в 63-й столбец
        row_values[62] = last_non_nan_value  # Перемещаем значение в последний столбец

    # Возвращаем обработанную строку в виде списка
    return row_values

# Применяем функцию ко всем строкам и конвертируем обратно в DataFrame
aligned_data = pd.DataFrame(rs_data.apply(move_last_non_nan_to_end, axis=1).tolist(), columns=rs_data.columns)

# Проверяем корректность данных
print("Проверяем данные после обработки:")
print(aligned_data.head(10))
print(aligned_data.info())

# Сохраняем итоговый DataFrame в Excel
aligned_data.to_excel('aligned_data_fixed.xlsx', index=False, engine='openpyxl')
print("Данные успешно сохранены в 'aligned_data_fixed.xlsx'")
```

```python

# Копируем данные в новый DataFrame для обработки
processed_data = aligned_data.copy()

# Функция для обработки Entry point
def parse_entry_point(value):
    if value == '0' or pd.isna(value):  # Если значение "0" или NaN
        return np.nan, np.nan
    restaurant = ''.join(filter(str.isdigit, str(value)))  # Извлекаем цифры
    city = ''.join(filter(str.isalpha, str(value)))  # Извлекаем буквы
    restaurant = int(restaurant) if restaurant else np.nan
    city = city if city else np.nan
    return restaurant, city

# Функция для обработки End point
def parse_end_point(value):
    if value == '-1' or pd.isna(value):  # Если значение "-1" или NaN
        return np.nan, np.nan
    restaurant = ''.join(filter(str.isdigit, str(value)))  # Извлекаем цифры
    restaurant = int(restaurant) if restaurant else np.nan
    action = 5
    return restaurant, action

# Обрабатываем столбцы Entry point и End point
processed_data[['Restaurant_entry', 'City_entry']] = processed_data['Entry point'].apply(
    lambda x: pd.Series(parse_entry_point(x))
)
processed_data[['Restaurant_end', 'Action_end']] = processed_data['End point'].apply(
    lambda x: pd.Series(parse_end_point(x))
)

# Удаляем исходные столбцы Entry point и End point
processed_data.drop(columns=['Entry point', 'End point'], inplace=True)

# Перемещаем Restaurant_entry и City_entry после столбца IP
columns_order = ['IP', 'Restaurant_entry', 'City_entry'] + [col for col in processed_data.columns if col not in ['IP', 'Restaurant_entry', 'City_entry']]
processed_data = processed_data[columns_order]

# Сохраняем результат в файл
processed_data.to_csv('processed_data.csv', index=False)
print("Таблица успешно сохранена в 'processed_data.csv'")
```

```python

# Максимальное количество ресторанов в данных
max_restaurants = sum(['Rated restaurant' in col for col in processed_data.columns])

# Создаем новый DataFrame для обработки
processed_data_final = pd.DataFrame()

# Итерация по строкам данных
for index, row in processed_data.iterrows():
    # Начало строки с уже существующими столбцами
    row_data = {
        'IP': row['IP'],
        'Restaurant_entry': row['Restaurant_entry'],
        'City_entry': row['City_entry'],
        'Restaurant_end': row['Restaurant_end'],
        'Action_end': row['Action_end']
    }
    action_index = 1  # Индекс для нумерации ресторанов

    # Обрабатываем рестораны и действия
    for i in range(1, max_restaurants + 1):
        restaurant_col = f'Rated restaurant{i}'

        if restaurant_col in row and not pd.isna(row[restaurant_col]):
            # Разделяем на ресторан (число) и действие (буква)
            restaurant_value = row[restaurant_col]
            if len(restaurant_value) > 1:
                restaurant = ''.join([c for c in restaurant_value if c.isdigit()])
                action = ''.join([c for c in restaurant_value if c.isalpha()])
            else:
                restaurant = restaurant_value  # Если только число
                action = 'No action'  # Действия нет

            row_data[f'Restaurant{action_index}'] = restaurant
            row_data[f'Action{action_index}'] = action
            action_index += 1

    # Добавляем обработанную строку
    processed_data_final = pd.concat([processed_data_final, pd.DataFrame([row_data])], ignore_index=True)

# Сохраняем результат
processed_data_final.to_csv('processed_data_final.csv', index=False)
print("Таблица успешно сохранена в 'processed_data_final.csv'")
```

```python
import pandas as pd

# Словарь для преобразования букв в рейтинги
rating_mapping = {
    'L': 1,  # Переключение (нейтральное)
    'M': 2,  # Поиск ресторана дешевле
    'N': 3,  # Поиск ресторана получше
    'O': 2,  # Поиск ближе
    'P': 4,  # Желание более традиционной кухни
    'Q': 4,  # Желание более креативной кухни
    'R': 3,  # Желание более оживлённой атмосферы
    'S': 3,  # Желание более тихой атмосферы
    'T': 1   # Желание смены кухни
}

# Создаем копию исходных данных
processed_data_original = processed_data_final.copy()

# Преобразование Action-столбцов, кроме Action_end
def map_action_to_rating(action):
    if pd.isna(action):  # Если NaN, оставляем его
        return action
    return rating_mapping.get(action, action)  # Преобразуем, если буква в словаре

# Обработка только столбцов 'Action' (кроме 'Action_end')
for column in processed_data_final.columns:
    if column.startswith('Action') and column != 'Action_end':
        processed_data_final[column] = processed_data_final[column].apply(map_action_to_rating)

# Перемещаем столбцы 'Restaurant_end' и 'Action_end' в конец
columns = list(processed_data_final.columns)
columns.remove('Restaurant_end')
columns.remove('Action_end')
columns += ['Restaurant_end', 'Action_end']
processed_data_final = processed_data_final[columns]


processed_data_final.to_excel('processed_data_final.xlsx', index=False)

print("- Обработанная версия: 'processed_data_final.xlsx'")
```
